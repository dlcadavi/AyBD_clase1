### Otros modos
Los 3 Modos B√°sicos (El "Sem√°foro" de archivos)

Existen otros "apodos" o modos:
'w' --> Write --> Escribe. Borra lo anterior y pone lo nuevo.
'r' --> Read --> Lee. Solo mira lo que hay, no deja cambiar nada.
'a' --> Append --> A√±ade. Escribe al final del archivo sin borrar lo que ya estaba.

```python
# Apend
# 1. El nuevo texto que queremos agregar
nuevo_texto = "\nEsta es una segunda l√≠nea que a√±adimos despu√©s."

# 2. Abrimos el archivo con 'a' (Append)
with open('hola.txt', 'a') as archivo:
    archivo.write(nuevo_texto)

print("‚úÖ Texto a√±adido al final del archivo.")
```

```python
'''
Atenci√≥n: para leer un archivo con 'r', el archivo ya debe existir en la carpeta de la izquierda.
'''

# 1. Abrimos el archivo con 'r' (Read)
with open('hola.txt', 'r') as archivo:
    # Guardamos lo que ley√≥ en una variable
    contenido_leido = archivo.read()

# 2. Mostramos el resultado en la consola
print("üìñ Esto es lo que dice el archivo actualmente:")
print("--------------------------------------------")
print(contenido_leido)
print("--------------------------------------------")
```

### 3.1: Listas de Datos

Las **Listas** son la estructura de datos m√°s b√°sica de Python. Son colecciones ordenadas que permiten almacenar varios elementos (nombres de candidatos, c√≥digos de zona, etc.).

**Recordatorio:** En Python, el conteo siempre comienza en el √≠ndice **cero (0)**.

```python
# ----------------------------------------------------
# Ejercicio 4.1: Listas y Acceso a Elementos
# ----------------------------------------------------

# 1. Cree una lista con los nombres de 5 ciudades de su regi√≥n
ciudades_principales = ["Medell√≠n", "Cali", "Barranquilla", "Cartagena", "Bucaramanga"]

# 2. Imprima la lista completa para verificar:
print(f"Lista completa: {ciudades_principales}")

# 3. Imprima el TERCER elemento de la lista
# (El tercer elemento est√° en el √≠ndice 2)
print(f"El tercer elemento es: {ciudades_principales[2]}")
```

### 4.2: Bucles (`for`) para Automatizaci√≥n

En la Semana 2, usaremos bucles `for` para automatizar tareas repetitivas, como **abrir y procesar los 100 archivos CSV** de la Registradur√≠a, uno por uno.

```python
# ----------------------------------------------------
# Ejercicio 4.2: Uso de Bucles 'for'
# ----------------------------------------------------

print("Iniciando simulaci√≥n de an√°lisis de ciudades:")

# Iteramos sobre la lista 'ciudades_principales'
for ciudad in ciudades_principales:
    # Por cada elemento, ejecutamos una acci√≥n
    print(f"-> Analizando resultados electorales de: {ciudad}")

print("\n¬°Proceso de iteraci√≥n finalizado!")
```

### 4.3: Diccionarios (Estructuras Clave - Valor)

Los **Diccionarios** son cruciales en anal√≠tica para almacenar metadatos donde necesitamos una **etiqueta (Clave)** para acceder r√°pidamente a un **valor** (ej. "Antioquia" nos da "Medell√≠n").

**Formato:** `{"Clave": "Valor", "Clave2": "Valor2"}`

```python
# ----------------------------------------------------
# Ejercicio 4.3: Acceso a Diccionarios
# ----------------------------------------------------

# Diccionario: Departamento y su capital (simulaci√≥n de metadato electoral)
departamentos_info = {
    "Antioquia": "Medell√≠n",
    "Valle del Cauca": "Cali",
    "Cundinamarca": "Bogot√°"
}

# Actividad: Imprima la capital de Antioquia
print(f"El metadato que buscamos (la capital de Antioquia) es: {departamentos_info['Antioquia']}")

# Reto Opcional: ¬øC√≥mo acceder√≠a usted a la capital del Valle del Cauca?
# print(f"La capital del Valle del Cauca es: {departamentos_info['Valle del Cauca']}")
```
